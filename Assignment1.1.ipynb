```python
def rth(r,g,b):
  r = r/255 
  g = g/255
  b = b/255
  Cmax = max(r,g,b)
  Cmin = min(r,g,b)
  dell= Cmax - Cmin
  if Cmax == Cmin:  
        h = 0
  elif Cmax == r:  
      h = (60 * ((g - b) / dell) + 0) % 360
  elif Cmax == g: 
      h = (60 * ((b - r) / dell) + 120) % 360
  elif Cmax == b: 
      h = (60 * ((r - g) / dell) + 240) % 360 

  if h < 0:
      h = h + 360
  if Cmax == 0: 
      s = 0
  else: 
      s = (dell / Cmax) * 100
      v = Cmax * 100

  a = round(h,2)
  b = round(s,2) 
  c = round(v,2)
  return a,b,c    
       
```

```python
print(rth(255,0,0))
```

```python
img = cv2.imread(r"C:\Users\mayur\OneDrive\Pictures\tulips.jpg")
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

lr = np.array([20,100,100])
ur = np.array([30,255,255])
mask2 = cv2.inRange(hsv,lr,ur)

lr = np.array([170,120,70])
ur = np.array([180,255,255])
mask1 = cv2.inRange(hsv,lr,ur)

lr = np.array([0,120,70])
ur = np.array([10,255,255])
mask3 = cv2.inRange(hsv,lr,ur)

lr = np.array([25,52,72])
ur = np.array([179,255,255])
mask4 = cv2.inRange(hsv,lr,ur)

mask = mask1 | mask2 | mask3 | mask4
res = cv2.bitwise_and(img,img,mask=mask)

k = img - res

plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(k[:,:,::-1])
plt.show()
```
